```r
library(fpp2)
library(ggfortify)
library(tidyverse)
library(kableExtra)
library(xts)
library(gridExtra)
library(ggResidpanel)
library(grid)
library(readr)
library(TTR)
library(ggpubr)
library(lattice)
library(dplyr)
```

```r
arrests <- read.table("https://raw.githubusercontent.com/NicJC/Datasets/main/Arrests.csv", 
                  header = TRUE,
                 sep = ",")
```

```r
head(arrests,15)
```

```r
plot(arrests$Year,arrests$Age, xlab = "Year", ylab = "Age" )
```

```r
ggplot(arrests, aes(Category, fill = Race, group = Race)) + 
  geom_bar()+ggtitle("Arrests")+ guides(x = guide_axis(angle = 90))+ scale_fill_discrete()
```

```r
ggplot(arrests, aes(Category,fill = Sex, group = Sex)) + geom_bar() +ggtitle("Category Arrests by Gender")+xlab("Category")+ylab("Arrests")+ guides(x = guide_axis(angle = 90))+ scale_fill_discrete()
```

```r
p <- ggplot(arrests) + 
  geom_density(aes(x = Year, fill = Sex), colour = NA) + 
  facet_grid(rows = vars(Race))

p
```

```r
boxplot(arrests$Age,horizontal=T,main="Age of Arrested",xlab="Age")
```
